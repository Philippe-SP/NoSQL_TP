```python
from pymongo import MongoClient

# Rcupration de la base de donnes et des collections
mongodb_client = MongoClient("localhost", 27017)
db = mongodb_client["AutoDrive_Pinheiro"]
timeseries_collection = db["timeseries"]

# Requete d'aggregation:
# Q5 : Dterminez les zones (bases sur des quadrants de coordonnes 
#      GPS) o les taxis sarrtent le plus frquemment dans une ville donne.

paris_zone = [[2.3522, 48.8566], [2.3522, 49.0000], [2.4000, 49.0000], [2.4000, 48.8566], [2.3522, 48.8566]] # coordonnes de Paris (environ)

pipeline = [
    {
        "$project": {
            "license_plate": 1,
            "lat": 1,
            "lon": 1,
            "status": 1,
            "location": { # Cration d'un champs location pour l'aggrgation
                "$let": { # Definition de variables locales pour la latitude et longitude
                    "vars": {
                        "longitude": "$lon",
                        "latitude": "$lat"
                    },
                    "in": [ # Ajout des variables dans un tableau
                        "$$longitude",
                        "$$latitude"
                    ]
                }
            }
        }
    },
    {
        "$match": { # Rcupration des taxis prsent dans la zone de Paris et stopp
            "status": "stopped",
            "location": {
                "$geoWithin": {
                    "$geometry": {
                        "type": "Polygon",
                        "coordinates": [paris_zone]
                    }
                }
            }
        }
    }
]

result = timeseries_collection.aggregate(pipeline)

# Affichage du resultat
for doc in result:
    print(doc)
```
